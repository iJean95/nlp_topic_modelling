```python
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sid = SentimentIntensityAnalyzer()

```

```python
#define function that detects sentiment
def sentiments(x):
    string = str(x)
    sentiment = sid.polarity_scores(string)['compound']
    return sentiment
```

```python
#apply sentiment of Cmt1
df["Comment_sentiment"] = df["Comment_translated"].apply(lambda x: sentiments(x))
```

```python
df.to_csv("sentiment_set.csv",sep=";")
```
